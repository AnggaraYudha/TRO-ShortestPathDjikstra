```python
def djikstra(graph, source, new_graph = None, visited_node = None, prev_dist = 0):
    if new_graph is None:
        new_graph = {}
    if visited_node is None:
        visited_node = []

    visited_node.append(source)
    if len(visited_node) == 1:
        new_graph[source] = {source:0}
    
    if (len(graph) == len(visited_node)):
        return new_graph
    
    min = 999
    next_node = ""
    for neighbor in graph[source]:
        node = next(iter(neighbor.keys()))
        dist = next(iter(neighbor.values())) + prev_dist
        
        if node in new_graph:
            if dist < next(iter(new_graph[node].values())):
                new_graph[node] = {source:dist}
        else:
            new_graph[node] = {source:dist}
    
    for node, origin in new_graph.items():
        dist = next(iter(origin.values()))
        if node not in visited_node:
            if dist < min:
                min = dist
                next_node = node

    return djikstra(graph, next_node, new_graph, visited_node, min)
```

```python
def back_track_from(graph, dest):
    cur = dest
    path = [dest]
    while cur != next(iter(graph[cur].keys())):
        cur = next(iter(graph[cur].keys()))
        path.append(cur)
    
    return path[::-1]
```

```python
import pandas as pd

pd.set_option('display.max_colwidth', None)

def output_table(graph, sources, destinations):
    header = {
        'Sumber': [],
        'Tujuan': [],
        'Path': [],
        'Cost': []
    }

    table = pd.DataFrame(header)
    table['Cost'] = table['Cost'].astype(int)

    for source in sources:
        new_graph = djikstra(graph, source)
        for dest in destinations:
            path = back_track_from(new_graph, dest)
            dist = next(iter(new_graph[dest].values()))
            
            # Buat path jadi string
            str_path = ""
            for node in path:
                str_path = str_path + node
                if node != dest:
                    str_path = str_path + " - "

            row = pd.DataFrame({'Sumber':source, 'Tujuan':dest, 'Path':[str_path], 'Cost':dist})
            table = pd.concat([table, row], ignore_index=True)

    return table
```
